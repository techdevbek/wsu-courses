     [ 2 -1  1][x1]   [ 2]

# Solution:

```python
# To solve Ax=b, We start by constructing the arrays for A and b.
import numpy as np
from numpy.linalg import eig

A = np.array([[ 2, -1,  1],
              [ 6,  1,  2],
              [ 2, -1,  3]])
b = np.transpose(np.array([ 2,  2,  0]))
# To solve the system we do

x = np.linalg.solve(A,b)
print("Real solution: ")
print(x)
```

The Jacobi Iteration applied to the given matrix A did not converge, as indicated by the solution vector [−2.7901,−936.0494,−55.0123], which shows magnitudes that are blowing up. This non-convergence occurs because the matrix
A is not well-preconditioned; specifically, partial pivoting has not been applied, meaning the largest absolute value in each column is not in the pivotal position.

To improve convergence, we will use partial pivoting by swapping rows 1 and 2 of A. This step will bring the largest absolute value in the first column to the top-left position, resulting in a preconditioned matrix that is better suited for iterative methods like the Jacobi Iteration.

After performing this row swap, we will proceed with the Jacobi Iteration on the new matrix, expecting it to yield a convergent solution due to the improved conditioning.
